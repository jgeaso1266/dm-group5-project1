```python
file_path_unperformed = "/Users/ylvaschuch/Desktop/Chopin/Sonata_2/4th/midi_score_annotations.txt"  
unperformed = [] 

with open(file_path_unperformed, 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        
        # Append the parts to the data list
        unperformed.append(float(columns[0]))

file_path_performed1 = "/Users/ylvaschuch/Desktop/Chopin/Sonata_2/4th/KaszoS16M_annotations.txt"
performed1 = []

with open(file_path_performed1, 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        # Append the parts to the data list
        performed1.append(float(columns[0]))
        
file_path_performed2 = "/Users/ylvaschuch/Desktop/Chopin/Sonata_2/4th/Knoll10_annotations.txt"
performed2 = []

with open(file_path_performed2, 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        # Append the parts to the data list
        performed2.append(float(columns[0]))
        
file_path_performed3 = "/Users/ylvaschuch/Desktop/Chopin/Sonata_2/4th/Giltburg02_annotations.txt"
performed3 = []

with open(file_path_performed3, 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        # Append the parts to the data list
        performed3.append(float(columns[0]))

#print("unperformed score has " + str(len(unperformed)) + " beats and performed score has " + str(len(performed)) + " beats")
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import FormatStrFormatter

fig, ax = plt.subplots(figsize=(12,12))
plt.plot(unperformed, performed1, 'orange', label='performed - Kaszo')
plt.plot(unperformed, performed2, 'seagreen', label='performed - Knoll')
plt.plot(unperformed, performed3, 'slateblue', label='performed - Giltburg')
plt.plot(unperformed, unperformed, 'firebrick', label='unperformed')
maxBins = 15
plt.gca().xaxis.set_major_locator(MaxNLocator(nbins=maxBins)) # nbins='auto' also possible
plt.gca().yaxis.set_major_locator(MaxNLocator(nbins=maxBins))
plt.xticks(rotation=90)
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
max_value = max(performed[-1], unperformed[-1])
plt.xlabel('Time (s)')
plt.ylabel('Performance (s)')
plt.legend()
plt.grid()
plt.title("Performed vs. unperformed version of Chopin's Sonata 2: 4th movement")
plt.show()
```

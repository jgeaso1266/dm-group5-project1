```python
file_path_unperformed = "scores/midi_score_annotations.txt"  
unperformed = [] 

with open(file_path_unperformed, 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        # Append the parts to the data list
        unperformed.append(float(columns[0]))

file_path_performed = "scores/ParkJH13M_annotations.txt"
performed = []

with open(file_path_performed, 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        # Append the parts to the data list
        performed.append(float(columns[0]))

print("unperformed score has " + str(len(unperformed)) + " beats and performed score has " + str(len(performed)) + " beats")
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import FormatStrFormatter

fig, ax = plt.subplots(figsize=(12,12))
plt.plot(unperformed, performed, 'g', label='performed')
plt.plot(unperformed, unperformed, 'r', label='unperformed')
maxBins = 15
plt.gca().xaxis.set_major_locator(MaxNLocator(nbins=maxBins)) # nbins='auto' also possible
plt.gca().yaxis.set_major_locator(MaxNLocator(nbins=maxBins))
plt.xticks(rotation=90)
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
max_value = max(performed[-1], unperformed[-1])
plt.xlabel('time (s)')
plt.ylabel('performance (s)')
plt.legend()
plt.title("Performed vs. unperformed version of C. Debussy's Reflets dans l'Eau (Images Book 1)")
plt.show()
```

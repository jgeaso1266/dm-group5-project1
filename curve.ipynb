```python
file_path = "scores/midi_score_annotations.txt"  
unperformed = [] 

with open(file_path, 'r') as file:
    for line in file:
        columns = line.strip().split('\t')
        # Append the parts to the data list
        unperformed.append(float(columns[0]))

print(len(unperformed))
```
